```python
import os
import json
import sys
from dotenv import load_dotenv
from IPython.display import Markdown, display, update_display

# Make sure we can import the week1/scraper.py module when running from repo root
sys.path.append(os.path.join(os.getcwd(), "week1"))
from scraper import fetch_website_contents, fetch_website_links
from openai import OpenAI

load_dotenv(override=True)
api_key = os.getenv("OPENAI_API_KEY")
if api_key and api_key.startswith("sk-proj") and len(api_key) > 10:
    print("got the key babyy!")
else:
    print("There might be a problem with your API key")

openai=OpenAI()
model='gpt-4.1-mini'


links=fetch_website_links("https://edwarddonner.com")
links
```

```python
link_system_prompt='''
    so you will be provided by various kinds of links of a website 
    and its your task to filter them and select all the relavent links we need for our website brochure builder
    you are a helpful and smart and to the point assistant and provide those links which you will filter not in markdown form
    respond to me in the json file format example-
    [
       'links':[
       {'type':'linkdin link','url':"url here"}
       {'type':' catagory','url':"url here"}
       ]
        
    ]'''


def fetch_links(url):
    user_prompt=f'''
    you are provided with various links of website {url}
    and you have to filter those links and find the relavent links for the company borchure 
    in the form "here are some relavent links for you:"
    '''
    links=fetch_website_links(url)
    user_prompt+='\n'.join(links)
    return user_prompt
    
    
print(fetch_links("https://edwarddonner.com"))
```

```python

def all_important_links(url):
    message=[
    {'role':'system','content':link_system_prompt},
    {'role':'user','content':fetch_links(url)}
    ]
    response=openai.chat.completions.create(model=model,messages=message,response_format={'type':'json_object'})
    result=response.choices[0].message.content
    links=json.loads(result)
    print(f'found {len(links['links'])} relavent links')
    return links

all_important_links("https://edwarddonner.com")
```

```python
#making of Brochure!!!!
def fetch_content_relavent_links(url):
    contents=fetch_website_contents(url)
    rel_links=all_important_links(url)
    result=f"### Website content {contents} ## Relavent links{rel_links}"
    for link in rel_links['links']:
        result+=f"links: {link['type']}"
        result+=fetch_website_contents(link['url'])
        return result

print(fetch_content_relavent_links("https://edwarddonner.com"))
```

```python
brochure_system_prompt='''
here all all the information and links for the website url we are providing
you are a helpful and on point assistant who will make a neat and nice brochure with proper markdown and headings and provide 
relavent data and helpful context links like facebook or to connect links like linkdin
'''

def brochure_user_prompt(company_name,url):
    user_prompt = f'''
    so you are provided with the company name {company_name} and the url of that company {url}
    you will be given all the relavemt data and links and you have to make sure to have it in structured and well organised format
    so a brochure can be made from those information
    '''
    content = fetch_content_relavent_links(url)
    user_prompt += content
    user_prompt = user_prompt[:5000]
    return user_prompt


print(brochure_user_prompt("ED Donner","https://edwarddonner.com"))
```

```python
company_name = "ED Donner"
url = "https://edwarddonner.com"

messages = [
    {'role': 'system', 'content': brochure_system_prompt},
    {'role': 'user', 'content': brochure_user_prompt(company_name, url)},
]
stream=openai.chat.completions.create(model=model,messages=messages,stream=True)
response=""
display_handle = display(Markdown(""), display_id=True)
for chunk in stream:
    response+=chunk.choices[0].delta.content or ''
    display_handle.update(Markdown(response))
    

```
